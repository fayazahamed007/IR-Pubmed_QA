```python
import string
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
stopwords_list = set(stopwords.words('english'))
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def my_tokenizer(doc):
    doc = doc.lower()    #Convert the document to lowercase
    words = word_tokenize(doc)    #Tokenize the document into individual words
    pos_tags = pos_tag(words)   #Perform POS tagging on the words
    
    non_stopwords = [w for w in pos_tags if not w[0].lower() in stopwords_list]   #Remove stopwords
    
    non_punctuation = [w for w in non_stopwords if not w[0] in string.punctuation]  #Remove punctuation
     
    lemmas_temp = []   # lemmatization based on the POS tags
    lemmas = []
    for w in non_punctuation:
        if w[1].startswith('J'):
            pos = wordnet.ADJ
        elif w[1].startswith('V'):
            pos = wordnet.VERB
        elif w[1].startswith('N'):
            pos = wordnet.NOUN
        elif w[1].startswith('R'):
            pos = wordnet.ADV
        else:
            pos = wordnet.NOUN
        
        lemmas_temp.append(lemmatizer.lemmatize(w[0], pos))
   
    for item in lemmas_temp:
        remove_repeats = (' '.join(OrderedDict.fromkeys(item.split())))
        lemmas.append(remove_repeats)

    return lemmas
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
def general(question):
    query_vect = tfidf_vectorizer.transform([question])
    similarity = cosine_similarity(query_vect, tfidf_matrix)
    index_similarity = np.argmax(similarity, axis=None)
    
    return similarity, index_similarity
```

```python
def ask(question):
    print('Your question:', question)
    
    similarity, index_similarity = general(question) 
    answer = df.iloc[index_similarity]['LONG_ANSWER']
    
    if answer == 'no':
        print("It's a No thing")
        
    elif answer == 'yes':
        print("It's a Yes thing")

    else:
        print("Maybe")
            
    if ((similarity[0, index_similarity])>0):
        print('CONTEXTS FOUND:', df.iloc[index_similarity]['CONTEXTS'])
        print('Similarity: {:.2%}'.format(similarity[0, index_similarity]))
        print('Answer:', df.iloc[index_similarity]['LONG_ANSWER'])
    else:
        print("Not really any great matches for that, sorry")
```
